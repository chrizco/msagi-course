# MSAGI RAG Pipeline Assignment

This notebook builds a Retrieval-Augmented Generation pipeline using PDF loading, chunking, OpenAI embeddings, FAISS vector storage, and semantic retrieval.

## Assignment goals
- Load 2-3 PDF documents
- Split documents into chunks
- Generate embeddings using the OpenAI API
- Store embeddings in a FAISS vectorstore
- Run semantic retrieval queries
- Report total chunks, average chunk size, and sample query output


## 1. Setup and Imports

This section imports the required libraries and defines the project paths.

```python
from pathlib import Path 
import os 
from dotenv import load_dotenv 

from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

PROJECT_ROOT = Path.cwd().parent
DATA_DIR = PROJECT_ROOT / "data" / "raw_pdfs"
VECTORSTORE_DIR = PROJECT_ROOT / "data" / "vectorstore"
OUTPUTS_DIR = PROJECT_ROOT / "outputs"

load_dotenv(PROJECT_ROOT / ".env")

print("Project root:", PROJECT_ROOT)
print("PDF data folder:", DATA_DIR)
print("Vectorstore folder:", VECTORSTORE_DIR)
print("Outputs folder:", OUTPUTS_DIR)
```

## 2. Check Selected PDFs

This section verifies that the selected PDFs are available in the assignment dataset folder.

```python
pdf_files = sorted(DATA_DIR.glob("*.pdf")) # sorted to ensure consistent order across runs; glob returns a generator, so we convert it to a list

print(f"Number of PDFs found: {len(pdf_files)}")
for pdf in pdf_files:
    print("-", pdf.name)

if len(pdf_files) < 2:
    raise ValueError("Expected at least 2 PDF files for the assignment.")
```

## 3. Load PDF Documents

This section loads the PDF files page by page using LangChain’s PyPDFLoader.

```python
documents = [] # empty list to hold all loaded documents

for pdf_path in pdf_files:
    loader = PyPDFLoader(str(pdf_path))
    pages = loader.load()
    documents.extend(pages) # extend the documents list with the pages from the current PDF
    print(f"Loaded {len(pages)} pages from {pdf_path.name}")

print(f"Total loaded pages: {len(documents)}")
```

## 4. Chunk the Documents

This section splits the loaded PDF pages into smaller text chunks. Chunking is necessary because long documents are too large to embed or retrieve efficiently as a whole.

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=150
)

chunks = text_splitter.split_documents(documents)

chunk_lengths = [len(chunk.page_content) for chunk in chunks]
average_chunk_size = sum(chunk_lengths) / len(chunk_lengths)

print(f"Total chunks: {len(chunks)}")
print(f"Average chunk size: {average_chunk_size:.0f} characters")
```

## 5. Inspect Example Chunks

This section displays a few example chunks to confirm that the chunking result is readable and meaningful.

```python
for i, chunk in enumerate(chunks[:3], start=1):
    print("=" * 80)
    print(f"Chunk {i}")
    print("Source:", Path(chunk.metadata.get("source", "unknown")).name)
    print("Page:", chunk.metadata.get("page", "unknown"))
    print(chunk.page_content[:1000])
```

## 6. Embeddings and FAISS Vectorstore

This section will generate embeddings with the OpenAI API and store them in a FAISS vectorstore.

Status: to be implemented after the PDF loading and chunking baseline works.

```python
# Create embeddings with OpenAI.
# Each text chunk is converted into a numerical vector.
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")

# Store all chunk vectors in a FAISS vectorstore.
vectorstore = FAISS.from_documents(chunks, embeddings)

print("FAISS vectorstore created.")
print("Number of embedded chunks:", len(chunks))
```

## 7. Retrieval Query

This section will run a sample user question and retrieve the most relevant chunks.

Status: to be implemented after FAISS setup works.

```python
# Sample question for semantic search.
query = "What does the document say about BPC-157 and wound healing?"

# Retrieve more chunks than we need, because some may be references or source lists.
raw_results = vectorstore.similarity_search(query, k=8)

# Keep only chunks that look like meaningful text.
results = []

for result in raw_results:
    text = result.page_content.strip()

    if len(text) > 400 and "references" not in text[:100].lower():
        results.append(result)

# Show the best 3 usable results.
results = results[:3]

print("Query:", query)
print("Number of usable retrieved chunks:", len(results))
print()

for i, result in enumerate(results, start=1):
    source_name = Path(result.metadata.get("source", "unknown")).name
    page_number = result.metadata.get("page", "unknown")

    print("=" * 80)
    print(f"Result {i}")
    print("Source:", source_name)
    print("Page:", page_number)
    print()
    print(result.page_content[:1500])
    print()
```

```python
# Save the FAISS vectorstore locally.
VECTORSTORE_DIR.mkdir(parents=True, exist_ok=True)

vectorstore.save_local(str(VECTORSTORE_DIR))

print("FAISS vectorstore saved to:")
print(VECTORSTORE_DIR)
```
